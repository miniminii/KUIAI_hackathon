```python
from webdriver_manager.chrome import ChromeDriverManager

# 인스타그램 로그인 화면 접속
driver=wd.Chrome(ChromeDriverManager().install())
driver.get("https://www.instagram.com/accounts/login/")
```

```python
# 로그인 아이디 입력
instagram_id_form = driver.find_element(By.NAME, "username")
instagram_id_form.send_keys("**인스타아이디**")
time.sleep(5)

# 로그인 비밀번호 입력
instagram_pw_form = driver.find_element(By.NAME, "password")
instagram_pw_form.send_keys("**인스타비밀번호")
time.sleep(7)

# 로그인 버튼 클릭     
login_ok_button = driver.find_element(By.XPATH, "//*[@id=\"loginForm\"]/div/div[3]/button")
login_ok_button.click()
time.sleep(5)
```

```python
# keyword로 인스타 인플루언서 계정 설정
keyword="skuukzky"
url = "https://www.instagram.com/{}/".format(keyword)
driver.get(url)
time.sleep(5)
```

```python
# 피드 갯수 크롤링
num_feed = int(driver.find_element(By.CLASS_NAME, "_ac2a").text)
print(num_feed)
first = driver.find_elements(By.CSS_SELECTOR, "div._aagw")[0]
first.click()
time.sleep(3)
```

```python
# 단일 피드 좋아요 갯수 크롤링
def get_num_like(dirver):
    try:
        html = driver.page_source
        soup = BeautifulSoup(html, "lxml")
        like = soup.select("div._aacl._aaco._aacw._aacx._aada._aade")[1].findAll('span')[-1].text
    except : 
        like = 0
    return like

# 다음 피드 넘어가기        
def move_next(driver):
    try:
        right = driver.find_element(By.CSS_SELECTOR,"div._aaqg._aaqh")
        right.click()
        time.sleep(3)
    except :
        print("finished")
```

```python
get_num_like(driver)
```

```python
num_like = []

#크롤링하려는 피드 갯수로 설정, 10은 테스트용
for i in range(num_feed):
    
    try : 
        data = get_num_like(driver)
        num_like.append(data)
        move_next(driver)
    # 좋아요 갯수 비공개 시 0 반환
    except:
        time.sleep(2)
        move_next(driver)
    time.sleep(5)
```
